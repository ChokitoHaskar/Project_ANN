```python
# Impor library
# Impor library yang dibutuhkan
from sklearn.neural_network import MLPClassifier as neuralNetwork
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import numpy as np
import joblib as jbl
```

```python
# Impor dataset
dataset_latih = pd.read_csv('./dataset_latih_final.csv')

# Acak urutan dataset
dataset_latih = dataset_latih.sample(frac = 1, random_state = 2025, ignore_index = True)
```

```python
# Tetapkan fitur dan label latih
fitur_latih = dataset_latih[['jenisKelamin', 'umur', 'beratBadan', 'tinggiBadan']]
target_latih = dataset_latih['label']
```

```python
# Buat model jaringan pertama sesuai dengan hasil cross-validation
modelAnn_kedua_1 = neuralNetwork(hidden_layer_sizes=(9,4),
                           activation='relu',
                           solver='adam',
                           batch_size=4,
                           learning_rate_init=0.01,
                           max_iter=600)

modelAnn_kedua_2 = neuralNetwork(hidden_layer_sizes=(9,4),
                           activation='relu',
                           solver='adam',
                           batch_size=4,
                           learning_rate_init=0.01,
                           max_iter=1000)

modelAnn_kedua_3 = neuralNetwork(hidden_layer_sizes=(9,4),
                           activation='relu',
                           solver='adam',
                           batch_size=4,
                           learning_rate_init=0.01,
                           max_iter=2200)

modelAnn_kedua_4 = neuralNetwork(hidden_layer_sizes=(9,4),
                           activation='relu',
                           solver='adam',
                           batch_size=8,
                           learning_rate_init=0.001,
                           max_iter=600)

modelAnn_kedua_5 = neuralNetwork(hidden_layer_sizes=(9,4),
                           activation='relu',
                           solver='adam',
                           batch_size=8,
                           learning_rate_init=0.001,
                           max_iter=1000)

modelAnn_kedua_6 = neuralNetwork(hidden_layer_sizes=(9,4),
                           activation='relu',
                           solver='adam',
                           batch_size=8,
                           learning_rate_init=0.001,
                           max_iter=2200)
```

```python
# Latih model jaringan
modelAnn_kedua_1.fit(fitur_latih, target_latih)
modelAnn_kedua_2.fit(fitur_latih, target_latih)
modelAnn_kedua_3.fit(fitur_latih, target_latih)
modelAnn_kedua_4.fit(fitur_latih, target_latih)
modelAnn_kedua_5.fit(fitur_latih, target_latih)
modelAnn_kedua_6.fit(fitur_latih, target_latih)
```

```python
# Impor dataset uji
dataset_uji = pd.read_csv('./dataset_uji_final.csv')

# Tetapkan fitur dan label uji
fitur_uji = dataset_uji[['jenisKelamin', 'umur', 'beratBadan', 'tinggiBadan']]
target_uji = dataset_uji['label']
```

```python
# Uji jaringan kedua
hasilAnn_kedua_1 = modelAnn_kedua_1.predict(fitur_uji)
hasilAnn_kedua_2 = modelAnn_kedua_2.predict(fitur_uji)
hasilAnn_kedua_3 = modelAnn_kedua_3.predict(fitur_uji)
hasilAnn_kedua_4 = modelAnn_kedua_4.predict(fitur_uji)
hasilAnn_kedua_5 = modelAnn_kedua_5.predict(fitur_uji)
hasilAnn_kedua_6 = modelAnn_kedua_6.predict(fitur_uji)
```

```python
# Cek performa seluruh tipe model pertama
print('Tipe 1 - Model kedua')
print(classification_report(target_uji, hasilAnn_kedua_1, zero_division=0))
print(confusion_matrix(target_uji, hasilAnn_kedua_1))

print('Tipe 2 - Model kedua')
print(classification_report(target_uji, hasilAnn_kedua_2, zero_division=0))
print(confusion_matrix(target_uji, hasilAnn_kedua_2))

print('Tipe 3 - Model kedua')
print(classification_report(target_uji, hasilAnn_kedua_3, zero_division=0))
print(confusion_matrix(target_uji, hasilAnn_kedua_3))

print('Tipe 4 - Model kedua')
print(classification_report(target_uji, hasilAnn_kedua_4, zero_division=0))
print(confusion_matrix(target_uji, hasilAnn_kedua_4))

print('Tipe 5 - Model kedua')
print(classification_report(target_uji, hasilAnn_kedua_5, zero_division=0))
print(confusion_matrix(target_uji, hasilAnn_kedua_5))

print('Tipe 6 - Model kedua')
print(classification_report(target_uji, hasilAnn_kedua_6, zero_division=0))
print(confusion_matrix(target_uji, hasilAnn_kedua_6))
```

```python
# Simpan seluruh tipe pada model jaringan kedua
jbl.dump(modelAnn_kedua_1, 'model_kedua_1.pkl')
jbl.dump(modelAnn_kedua_2, 'model_kedua_2.pkl')
jbl.dump(modelAnn_kedua_3, 'model_kedua_3.pkl')
jbl.dump(modelAnn_kedua_4, 'model_kedua_4.pkl')
jbl.dump(modelAnn_kedua_5, 'model_kedua_5.pkl')
jbl.dump(modelAnn_kedua_6, 'model_kedua_6.pkl')
```
