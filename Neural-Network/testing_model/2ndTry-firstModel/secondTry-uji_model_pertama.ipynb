```python
from sklearn.neural_network import MLPClassifier as ANN
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import joblib as jbl
```

```python
dataset_latih = pd.read_csv('./dataset_latih_final.csv')
dataset_latih.head(10)
```

```python
dataset_uji = pd.read_csv('./dataset_uji_final.csv')
dataset_uji.head(10)
```

```python
dataset_latih.value_counts('label', sort = False)
```

```python
dataset_uji.value_counts('label', sort = False)
```

```python
fitur_latih = dataset_latih[['jenisKelamin', 'umur', 'beratBadan', 'tinggiBadan']]
target_latih = dataset_latih['label']
```

```python
fitur_uji = dataset_uji[['jenisKelamin', 'umur', 'beratBadan', 'tinggiBadan']]
target_uji = dataset_uji['label']
```

```python
model_pertama_tipe1 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 2,
                          learning_rate_init = 0.001,
                          max_iter = 600
                         )

model_pertama_tipe2 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 2,
                          learning_rate_init = 0.01,
                          max_iter = 600
                         )

model_pertama_tipe3 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 2,
                          learning_rate_init = 0.001,
                          max_iter = 2200
                         )

model_pertama_tipe4 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 2,
                          learning_rate_init = 0.01,
                          max_iter = 2200
                         )

model_pertama_tipe5 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 8,
                          learning_rate_init = 0.001,
                          max_iter = 600
                         )

model_pertama_tipe6 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 8,
                          learning_rate_init = 0.01,
                          max_iter = 600
                         )

model_pertama_tipe7 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 8,
                          learning_rate_init = 0.001,
                          max_iter = 2200
                         )

model_pertama_tipe8 = ANN(hidden_layer_sizes = (9,),
                          activation = 'relu',
                          solver = 'adam',
                          batch_size = 8,
                          learning_rate_init = 0.01,
                          max_iter = 2200
                         )
```

```python
model_pertama_tipe1.fit(fitur_latih, target_latih)
model_pertama_tipe2.fit(fitur_latih, target_latih)
model_pertama_tipe3.fit(fitur_latih, target_latih)
model_pertama_tipe4.fit(fitur_latih, target_latih)
model_pertama_tipe5.fit(fitur_latih, target_latih)
model_pertama_tipe6.fit(fitur_latih, target_latih)
model_pertama_tipe7.fit(fitur_latih, target_latih)
model_pertama_tipe8.fit(fitur_latih, target_latih)
```

```python
hasil_tipe1 = model_pertama_tipe1.predict(fitur_uji)
hasil_tipe2 = model_pertama_tipe2.predict(fitur_uji)
hasil_tipe3 = model_pertama_tipe3.predict(fitur_uji)
hasil_tipe4 = model_pertama_tipe4.predict(fitur_uji)
hasil_tipe5 = model_pertama_tipe5.predict(fitur_uji)
hasil_tipe6 = model_pertama_tipe6.predict(fitur_uji)
hasil_tipe7 = model_pertama_tipe7.predict(fitur_uji)
hasil_tipe8 = model_pertama_tipe8.predict(fitur_uji)
```

```python
print(classification_report(target_uji, hasil_tipe1, zero_division=0))
print(confusion_matrix(target_uji, hasil_tipe1))
```
